# Filtering 

```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
x = np.linspace(1,20,100)
y = np.cos(x)
```

```python
plt.plot(x,y)
plt.title("Plot of f(x)")
plt.show()
```

### Adding some noise to our original signal 

```python
noise = np.random.randn(100)

noisy_signal = y+ 0.1*noise
```

```python
plt.plot(x, noisy_signal)
plt.title("Noise signal")
plt.show()
```

```python
convolved_signal = np.convolve(noisy_signal, np.ones(5)/5)
```

```python
plt.plot(x, convolved_signal[2:-2])
plt.show()
```

##  Lets try apply this in real time ! 

```python
import cv2
```

```python
image = cv2.imread("./imgs/download.jfif")
images = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(8,8))
plt.imshow(image)
plt.show()
```

```python
plt.figure(figsize=(15,15))

for s, i in enumerate([3,7,11,25]):
    mean_filter = cv2.filter2D(images,-1, np.ones((i,i))/i**2 )
    plt.title("Mean Filter")
    plt.subplot(2,2,s+1)
    plt.imshow(mean_filter)

plt.show()
```

## Gaussina Filter 

```python
plt.figure(figsize=(15,15))

for s,i in enumerate([3,7,11,25]):
    
    img = cv2.GaussianBlur(image, (i,i), i/6)
    
    plt.title(f"Gaussian filter {s}")
    plt.subplot(3,3,s+1)
    plt.imshow(img)  

plt.show()
```

## Mean Filter 

```python
plt.figure(figsize= (10,10))
img = cv2.blur(image, (11,11))
plt.imshow(img)
plt.title("Mean Filter")
plt.show()
```

## Median Filter

```python
plt.figure(figsize= (10,10))
img = cv2.medianBlur(image, 11)
plt.imshow(img)
plt.title("Median Filter")
plt.show()
```

## Gaussian Filter

```python
plt.figure(figsize= (10,10))
img = cv2.GaussianBlur(image, (11,11), 11/6)
plt.imshow(img)
plt.title("Gaussian Filter")
plt.show()
```
